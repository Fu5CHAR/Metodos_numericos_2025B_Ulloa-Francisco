**2. Determine cuáles de las siguientes matrices son no singulares y calcule la inversa de esas matrices**
$$
a. \begin{bmatrix}
&4 &2 &6\\
&3 &0 &7\\
&-2 &-1 &-3
\end{bmatrix}

b. \begin{bmatrix}
&1 &2 &0\\
&2 &1 &-1\\
&3 &1 &1
\end{bmatrix}
\\\\
$$
$$
\\\\
c. \begin{bmatrix}
&1 &1 &-1 &1\\
&1 &2 &-4 &-2\\
&2 &1 &1 &5\\
&-1 &0 &-2 &-4
\end{bmatrix}

d. \begin{bmatrix}
&4 &0 &0 &0\\
&6 &7 &0 &0\\
&9 &11 &1 &0\\
&5 &4 &1 &1
\end{bmatrix}
$$

```python
import numpy as np
A=np.array([[4,2,6],[3,0,7],[-2,-1,-3]])
B=np.array([[1,2,0],[2,1,-1],[3,1,1]])
C=np.array([[1,1,-1,1],[1,2,-4,-2],[2,1,1,5],[-1,0,2,-4]])
D=np.array([[4,0,0,0],[6,7,0,0],[9,11,1,0],[5,4,1,1]])
def invert_matrix(matrix):
    if np.linalg.det(matrix) != 0:
        inv_matrix = np.linalg.inv(matrix)
        return print('La matriz :\n', matrix, '\nes no singular y la inversa de la matriz es:\n', inv_matrix)
    else:
        return print("La matriz:\n", matrix, "\nes singular y no tiene inversa.")
    
invert_matrix(A)
invert_matrix(B)
invert_matrix(C)
invert_matrix(D)
```

**4. Encuentre los valores de A que hacen que la siguiente matriz sea singular.**
$$
A= 
\begin{bmatrix}
&1 & -1 & \alpha\\
&2 &2 &1\\
&0 &\alpha & \frac{-3}{2}
\end{bmatrix}
$$

Para que una matriz sea singular, esta no debe ser invertible, es decir, su determinante debe ser igual a cero.<br>
Entonces buscamos los valores de $\alpha$ tal que $det[A]=0$.<br>
Luego,

\begin{equation*}
\begin{split}
det[A] &= ((1 * 2 *\frac{-3}{2})+(2 * \alpha * \alpha)+(0 * 1 * -1)) -((\alpha * 2 * 0)+(1 * \alpha * 1)+(\frac{-3}{2} * 2 * -1))\\
&= (-3+2(\alpha)^2 + 0) - (0 + \alpha + 3)\\
&= 2(\alpha)^2 - \alpha - 6
\end{split}
\end{equation*}
<br>

Si $det(A)=0$, entonces<br>

\begin{equation*}
\begin{split}
2(\alpha)^2 - \alpha - 6 &=0\\
(\alpha -3)(\alpha +2) &= 0
\end{split}
\end{equation*}
<br>

Finalmente, la matriz A es singular para $\alpha=3$ y $\alpha=-2$.
