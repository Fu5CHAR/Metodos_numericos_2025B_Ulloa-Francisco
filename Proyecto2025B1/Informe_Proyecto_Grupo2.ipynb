<h3 style="text-align:center;">Desarrolo matematico</h3>

<h3 style="text-align:center;"> Inter茅s compuesto</h3>



El inter茅s compuesto se genera cuando los intereses generados en un per铆odo se suman al capital base, de modo que, en el siguiente per铆odo, se calculan nuevos intereses sobre el capital incrementado. Este proceso se repite peri贸dicamente, provocando que el capital crezca de manera acelerada en comparaci贸n con el inter茅s simple. La frecuencia con la que se generan estos intereses (mensual, trimestral, anual) afecta al capital final acumulado, ya que a mayor frecuencia de capitalizaci贸n, mayor ser谩 el efecto acumulado.


<h3 style="text-align:center;"> Depositos continuos</h3>


En los dep贸sitos continuos, el dinero se agrega al capital de la cuenta de manera constante, en lugar de realizar dep贸sitos peri贸dicos en intervalos fijos como ocurre en los dep贸sitos mensuales. El inter茅s compuesto continuo se describe con la f贸rmula:

$A=Pe^rt$
 
En esta f贸rmula, el inter茅s se calcula de manera continua sobre el saldo acumulado, sin esperar a un per铆odo de capitalizaci贸n espec铆fico. Esto permite una acumulaci贸n m谩s fluida del inter茅s a lo largo del tiempo. 

<h3 style="text-align:center;"> Metodo de la Bisecci贸n</h3>


El m茅todo de la bisecci贸n se utiliza para encontrar la ra铆z de una funci贸n. En este escenario, lo implementamos para calcular la tasa de inter茅s compuesto con aportes peri贸dicos. Esto es necesario porque, para calcular la tasa de inter茅s $i$ cuando conocemos el valor final $V_f$, el valor inicial $V_o$, y el n煤mero de per铆odos $n$, necesitamos resolver la siguiente ecuaci贸n:

$V_f = V_o (1+i)^n$ 
  
Debido a que trabajamos con una ecuaci贸n no lineal, no podemos resolverla de manera algebraica directa para $i$. Dado este escenario, utilizamos el m茅todo de bisecci贸n, que consiste en encontrar un intervalo en el que la funci贸n cambie de signo. El m茅todo divide repetidamente el intervalo a la mitad y eval煤a la funci贸n en los puntos medios, ajustando el intervalo hasta encontrar una soluci贸n con la precisi贸n requerida.

Para implementar la calculadora de intereses, es necesario incluir una validaci贸n de los datos ingresados, asegur谩ndonos de que tanto la tasa de inter茅s como los per铆odos sean valores positivos. Adem谩s, debemos considerar la precisi贸n de los c谩lculos, ya que, en ocasiones, el valor de $i$ puede converger lentamente o tener una precisi贸n limitada debido al comportamiento del m茅todo de bisecci贸n.

1) Modelo y convenci贸n 

Hay n periodos (por ejemplo n semanas).El dep贸sito de la semana 1 es V0 (es decir, el primer aporte = V0).
A partir de la semana 2 y hasta la n, cada semana aportas la misma cantidad A (constante).La secuencia de aportes es:
 A1 = V0, A2 = A, A3 = A, ..., An = A.

El inter茅s i es la tasa peri贸dica por semana (lo que queremos encontrar).El procedimiento que refleja tu tabla: en cada semana primero sumas el aporte y luego se aplica inter茅s.Con esto, la recurrencia es:

$0=0$

$S_t= (S_t-1  + A_t)(1+i)$

para t = 1,2,....,n

y el valor final (al terminar periodo n) es

VF = $S_n$

2) Forma cerrada (suma expl铆cita)

Desplegando la recurrencia se obtiene:

$S_n=\sum_{k=1}^{n}A_k(1+i^n-k+1.)$


Con tu patr贸n de aportes $A_1 = V_0 , A_k = A(k>=2):$


$VF(i)=V_0(1+i)^n + \sum_{k=2}^{n} * A_k(1+i^n-k+1.)$

la suma geom茅trica puede simplificarse si $i$ es diferente 0 a:

$VF(i) = V_0 (1+i)^n + A*(\frac{((1+i)^n)-(1+i)}{i})*(1+i) $

	
3) Planteamiento para encontrar la tasa i

Si conoces VF_dado (valor final observado), V0, A, n, defines la funci贸n objetivo:

$F(i)=VF(i)-VF_dado$

queremos encontrar la ra铆z $i$ tal que $f(i)=0$


<h3 style="text-align:center;">Descripci贸n de la implementaci贸n, resaltando el m茅todo num茅rico utilizado</h3>

La calculadora de tasa de inter茅s anual realiza un enfoque num茅rico para calcular la tasa de inter茅s necesaria para alcanzar un valor final espec铆fico en una cuenta de ahorros, comenzando con el valor del dep贸sito inicial, los aportes peri贸dicos y la duraci贸n del per铆odo de inversi贸n.

en la resolucion del problema usamos la formula para calcular el valor final $V_f$ con inter茅s compuesto y aportes peri贸dicos 

$V_f = V_o * (1+i)^n + A* ((1+i)^n  -1 / i)$

donde 
- $V_f$: valor final esperado.
- $V_o$: valor inicial del deposito.
- $A$: aporte periodico. 
- $i$: tasa de inter茅s por per铆odo.
- $n$: n煤mero de per铆odos de inversi贸n.

* Metodo de Bisecci贸n 

Es un algoritmo que busca la ra铆z de una funci贸n en un intervalo. Para hallar la tasa de inter茅s $i$ formulamos la funci贸n $g(i)$, que representa la diferencia entre el valor final calculado con una tasa $i$ y el valor deseado $V_f$.

$g(i) = V_o * (1+i)^n + A ((1+i)^n - 1 /i) - V_f$

el objetivo es encontrar el valor de $i$ tal que $g(i)=0$

* implementacion del metodo de Bisecci贸n 

en nuestro c贸digo el algoritmo de la bisecci贸n sigue los siguientes pasos:

* 1. Inicializaci贸n de valores: Asignamos un intervalo inicial para la tasa de inter茅s, que puede ser positivo o negativo. Esto depende de la comparaci贸n inicial entre $ g(0) y V_f$. si $g(0)$ es menor que cero, buscamos una tasa postiva; si es mayor que cero buscamos una tasa negativa. 

* 2. Evaluaci贸n de la funci贸n: En cada iteraci贸n, se eval煤a la funci贸n $g(i)$ en el punto medio del intervalo y se ajusta el intervalo dependiendo del signo de $g(i)$

    - si $g(i)$ es negativo, significa que el valor final calculado es menor que $V_f$, por lo que se ajusta en el intervalo superior.

    - so $g(i)$ es positivo el valor final calculado es mayor que $V_f$ por lo que se ajusta el intervalo inferior.

* 3. Criterio de parada: El algoritmo contin煤a iterando hasta que la diferencia entre los extremos sea menor que un valor de tolerancia o hasta que la funci贸n evaluada en el punto medio sea lo suficientemente cercana a cero.

* 4.Resultado: cuando el algoritmo finaliza, el valor de $i$ es el punto medio del intevalo final, que representa la tasa de inter茅s que permite alcanzar el valor final esperado. 







# An谩lisis anal铆tico de estabilidad y convergencia del m茅todo implementado

An谩lisis de la Convergencia en el M茅todo de Bisecci贸nEl m茅todo de bisecci贸n se caracteriza por una convergencia lineal, lo que implica que el error se reduce a la mitad en cada iteraci贸n.

Aunque su velocidad es inferior a la de algoritmos de segundo orden como Newton-Raphson, su principal ventaja radica en su robustez: garantiza encontrar la soluci贸n siempre que la ra铆z se encuentre dentro del intervalo inicial. 

Esto lo hace id贸neo para despejar variables complejas, como tasas de inter茅s impl铆citas.Desde el punto de vista computacional, la precisi贸n aumenta proporcionalmente al n煤mero de iteraciones ($N$), mejorando en un factor de 2 por ciclo.
 
El algoritmo finaliza cuando la amplitud del intervalo o el valor de la funci贸n es menor a una tolerancia preestablecida (por ejemplo, $\epsilon = 10^{-6}$).

### FUNCIN find_rate_bisection(V0, A, n, Vf, limite_inf=0.0, limite_sup=1.0, tolerancia=1e-12, max_iter=200)
```
INICIO
    INTENTAR
        fa = f(limite_inf, V0, A, n, Vf)
        fb = f(limite_sup, V0, A, n, Vf)
    SI_ERROR
        RETORNAR NULL
    FIN_INTENTAR
    
    // Expandir l铆mite superior si no hay cambio de signo
    contador_expansion = 0
    MIENTRAS (fa  fb > 0) Y (contador_expansion < 50) HACER
        limite_sup = limite_sup  2
        fb = f(limite_sup, V0, A, n, Vf)
        contador_expansion = contador_expansion + 1
    FIN_MIENTRAS
    
    SI fa  fb > 0 ENTONCES
        RETORNAR NULL  // No se encontr贸 cambio de signo
    FIN_SI
    
    // M茅todo de bisecci贸n
    punto_medio = NULL
    PARA iteracion DESDE 1 HASTA max_iter HACER
        punto_medio = (limite_inf + limite_sup) / 2
        fm = f(punto_medio, V0, A, n, Vf)
        
        SI |fm| < tolerancia O (limite_sup - limite_inf)/2 < tolerancia ENTONCES
            RETORNAR punto_medio
        FIN_SI
        
        SI fa  fm < 0 ENTONCES
            limite_sup = punto_medio
            fb = fm
        SINO
            limite_inf = punto_medio
            fa = fm
        FIN_SI
    FIN_PARA
    
    RETORNAR punto_medio
FIN
```
