### Black
Black es un formatter **opinionated**: toma todas las decisiones por ti y siempre produce el mismo resultado.
- Se instala f치cilmente con `pip install black`.
- Tiene extensi칩n oficial para VS Code.
- Permite formatear archivos completos, carpetas y proyectos.

### autopep8
autopep8 es un formatter m치s **flexible**, orientado a corregir violaciones del est치ndar PEP 8.
- Instalaci칩n sencilla con `pip install autopep8`.
- Muy bien integrado con VS Code.
- Permite formatear archivos completos **o solo fragmentos de c칩digo**.

### YAPF
YAPF (Yet Another Python Formatter) fue creado por Google.
- Se instala con `pip install yapf`.
- Solo se usa para formatear archivos o proyectos completos.
- Menos integraci칩n directa con formateo por selecci칩n en editores.
- Puede producir resultados distintos seg칰n la configuraci칩n elegida.

```python
##Con black
import numpy as np
import matplotlib.pyplot as plt


def jacobi_modificado(A, b, X0, TOL=1e-6, N=100):
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    X0 = np.array(X0, dtype=float)

    n = len(b)

    # Historial de iteraciones (iteraci칩n 0)
    tabla = [X0.copy()]

    # 游댳 Verificaci칩n previa (iteraciones = 0)
    if np.all(np.abs(A @ X0 - b) < TOL):
        print("N칰mero de iteraciones: 0")
        return np.array(tabla)

    k = 1

    while k <= N:
        x = np.zeros(n)

        # Paso 3
        for i in range(n):
            suma = 0.0
            for j in range(n):
                if j != i:
                    suma += A[i, j] * X0[j]

            x[i] = (b[i] - suma) / A[i, i]

        tabla.append(x.copy())

        # Paso 4: condici칩n componente a componente
        if np.all(np.abs(x - X0) < TOL):
            print("N칰mero de iteraciones:", k - 1)
            return np.array(tabla)

        X0 = x.copy()
        k += 1

    return "N칰mero m치ximo de iteraciones excedido"
```
