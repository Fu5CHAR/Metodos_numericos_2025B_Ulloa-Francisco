```python
import numpy as np
import matplotlib.pyplot as plt

def jacobi(A, b, X0, TOL=1e-6, N=100):

    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    X0 = np.array(X0, dtype=float)

    n = len(b)
    k = 1

    # Historial de iteraciones
    tabla = [X0.copy()]

    while k <= N:
        x = np.zeros(n)

        # Paso 3
        for i in range(n):
            suma = 0.0
            for j in range(n):
                if j != i:
                    suma += A[i, j] * X0[j]

            x[i] = (b[i] - suma) / A[i, i]

        # Guardar iteraciÃ³n completa
        tabla.append(x.copy())

        # Paso 4: condiciÃ³n componente a componente
        if np.all(np.abs(x - X0) < TOL):
            tabla = np.array(tabla)

            print('NÃºmero de iteraciones:', k)
            print(tabla)
            return tabla

        # Paso 5 y 6
        X0 = x.copy()
        k += 1

    return "NÃºmero mÃ¡ximo de iteraciones excedido"
```

```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
A =[[7,1,1],[-3,7,-1],[-2,5,9]]
b=[6,-26,1]
x0=[0,0,0]
tabla= jacobi(A,b,x0)


#GrÃ¡fica 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(tabla[:, 0], tabla[:, 1], tabla[:, 2],marker='o', linestyle='-')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Convergencia del mÃ©todo de Jacobi')
plt.show()
 
```

```python
import numpy as np
import matplotlib.pyplot as plt

def jacobi_modificado(A, b, X0, TOL=1e-6, N=100):

    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    X0 = np.array(X0, dtype=float)

    n = len(b)

    # Historial de iteraciones (iteraciÃ³n 0)
    tabla = [X0.copy()]

    # ðŸ”¹ VerificaciÃ³n previa (iteraciones = 0)
    if np.all(np.abs(A @ X0 - b) < TOL):
        print('NÃºmero de iteraciones: 0')
        return np.array(tabla)

    k = 1

    while k <= N:
        x = np.zeros(n)

        # Paso 3
        for i in range(n):
            suma = 0.0
            for j in range(n):
                if j != i:
                    suma += A[i, j] * X0[j]

            x[i] = (b[i] - suma) / A[i, i]

        tabla.append(x.copy())

        # Paso 4: condiciÃ³n componente a componente
        if np.all(np.abs(x - X0) < TOL):
            print('NÃºmero de iteraciones:', k - 1)
            return np.array(tabla)

        X0 = x.copy()
        k += 1

    return "NÃºmero mÃ¡ximo de iteraciones excedido"
```

```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
S1 = [[1,1],[-2,5]]
b1=[7,0]
x02=[5,2]
jacobi_modificado(S1,b1,x02)
tabla=jacobi_modificado(S1,b1,x02)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(tabla[:, 0], tabla[:, 1],marker='o', linestyle='-')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Convergencia del mÃ©todo de Jacobi')
plt.show()
```
