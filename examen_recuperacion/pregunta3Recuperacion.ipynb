```python
def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3-9.125*x**2-14.625*x-5.625

import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(-3,3,100)
y= f(x)

plt.plot(x,y)
plt.show()



```

```python
from scipy import optimize

root = optimize.newton(f, 0)

print(root)
```
