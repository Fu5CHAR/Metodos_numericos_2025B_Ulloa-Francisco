**Algoritmo 1**

```python
def suma_numeros(lista):
    SUM = 0
    for x in lista:
        SUM = SUM + x
    return SUM


def main():
    numeros = [1, 1/2, 1/4, 1/8, 1/16, 1/32]
    resultado = suma_numeros(numeros)
    print(f"La suma de los números es: {resultado}")

    x = abs((resultado - 2) / 2) * 100
    print(f"El error es: {x}%")

    
if __name__ == "__main__":
    main()
```

**Algoritmo 2**

```python
def bubbleSort(array):
    n = len(array)
    comparaciones = 0

    print("\nProceso de ordenamiento:")
    for i in range(n):
        swapped = False
        print(f"Iteración {i + 1}:")
        for j in range(1, n - i):
            comparaciones += 1
            if array[j] < array[j - 1]:
                array[j], array[j - 1] = array[j - 1], array[j]
                swapped = True
            print(f"  Comparando índice {j-1} y {j}: {array}")
        if not swapped:
            print("  No hubo intercambios, la lista ya está ordenada.\n")
            break
        print(f"  Resultado después de la iteración {i + 1}: {array}\n")

    return array, comparaciones


def main():
    lista = [, , , , ]
    print(f"Lista original: {lista}")

    lista_ordenada, num_comparaciones = bubbleSort(lista)

    print(f"\nLista ordenada: {lista_ordenada}")
    print(f"Número total de comparaciones: {num_comparaciones}")


if __name__ == "__main__":
    main()
```

**Algoritmo 3**


```python
def iterative(n):
    if n <= 0:
        return 0
    else:
        x = 0
        y = 1
        for i in range(1, n):
            z = x + y
            x = y
            y = z
        return y


def main():
    valores = [11, 84, 1581]

    for n in valores:
        resultado = iterative(n)
        print(f"n = {n} → y = {resultado}")


if __name__ == "__main__":
    main()
```
